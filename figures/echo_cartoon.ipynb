```python
xlist = np.linspace(0, max(tlist)-.005, 1000)
tau = .03

vals = np.zeros(len(xlist))

vals[1] = 13
for i in range(2, 400):
    vals[i] = 10*np.exp(-(xlist[i]**2)/(.00005))


vals[np.argmin(np.abs(xlist-tau))] = 13

for i in range(500,len(xlist)):
    vals[i] = 4*np.exp(-(xlist[i]-2*tau)**2/(.00005))

plt.plot(xlist, vals)
plt.ylim([-4,23])
plt.xlim([-.01,.08])
plt.ylabel(r"$M_{xy} (V)$")
plt.xlabel(r"$t (s)$")

plt.yticks([])
plt.xticks([])

plt.gcf().set_dpi(200)
```
